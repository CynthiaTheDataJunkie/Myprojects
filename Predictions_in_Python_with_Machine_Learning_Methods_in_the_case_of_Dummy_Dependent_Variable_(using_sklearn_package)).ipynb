The Interquartile Range (IQR) is a method to detect and remove outliers from a dataset. The IQR is the range between the first quartile (25th percentile) and the third quartile (75th percentile) of the data. It represents the central 50% of the data and is a measure of data dispersion. The IQR method can be used to identify and remove outliers because data points outside the IQR are considered extreme values and could distort the analysis.
After removing the outliers, you can retrain your models on the cleaned dataset and evaluate their performance again. Keep in mind that removing outliers can improve the performance of some models, but it might not always result in significant improvements. 

```python
# define a function to remove outliers using IQR
def remove_outliers_iqr(df, columns):
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# specify the columns to remove outliers from
columns_to_check = ['size', 'Revenue']

# call the function to remove outliers using IQR
df_clean = remove_outliers_iqr(df_innovativenessNew, columns_to_check)

# print the resulting dataframe
df_clean.head()
```

## TASK 3C LOGISTIC REGRESSION


Logistic Regression is a statistical method and supervised machaine learning method used for analyzing a dataset in which there are one or more independent variables that determine an outcome. The outcome is measured with a dichotomous variable (0 or 1). It is used to predict a binary outcome (Yes/No, True/False, Success/Failure, etc.) based on one or more predictor (independent) variables. Logistic Regression uses the logistic function, which is a sigmoid function, to model the probability of the binary outcome given the predictor variables.The logistic function is represented as:

P(Y=1 | X) = 1 / (1 + e^(-(b0 + b1 * X1 + b2 * X2 + ... + bn * Xn)))

**The Logistic function for this research is**

**P(Innovativeness=1 | X) = 1 / (1 + e^(-(b0 + b1 * Revenue + b2 * Size + β3 * Service Launch +  β4 * r&d + β5 * Investment + β6 * training + β7 * bonussystem )))**

```python
import statsmodels.api as sm #Stasmodels package gives values of parameters (log-odds ratios, odds ratios, marginal effects).

import numpy as np
df_clean['constant'] = 1 # statsmodel package requires constant term in X matrix; skmodel package has built in constant
X = df_clean[['constant','Service Launch','r&d','investment','size','training','bonus system','Revenue']]
innovativeness = df_clean[['innovativeness']]  #independent variables
logit_model=sm.Logit(innovativeness,X)
result=logit_model.fit()
print("Log-odds ratio")
print(result.summary2())

print("Odds ratio")
print(round(np.exp(result.params), 3))  #odds-ratio

AME = result.get_margeff(at = "overall", method = "dydx")  #marginal effects
print("Marginal Effects")
print(AME.summary())
```

```python
from sklearn.metrics import roc_curve, auc

# calculate predicted probabilities
y_prob = rf.predict(X_test)

# calculate false positive rate, true positive rate, and threshold values
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# calculate area under the curve (AUC)
rf_auc = auc(fpr, tpr)

# plot ROC curve
plt.plot(fpr, tpr, label='Random Forest (AUC = {:.2f})'.format(rf_auc))
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

# An AUC of 0.86 suggests that your Random Forest model can accurately distinguish between high and low employee innovativeness with a 86% probability, which is an excellent level of predictive performance. This demonstrates that the Random Forest classifier is a strong candidate for predicting employee innovativeness in this context, compared to the logistic regression model with an AUC of 70%.

**ANN stands for Artificial Neural Network, which is a computational model inspired by the structure and functioning of biological neural networks, such as those found in the human brain. ANNs are a type of machine learning model used for a wide range of tasks, including classification, regression, pattern recognition, and natural language processing.**

**Using the Multi-layer Perceptron (ANN) classifier to create an Artificial Neural Network model and using the Root mean square error, Mean Square Error and other relevant performance metrics to evaluate the performance of the model**

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# initialize ANN classifier
ann = MLPClassifier()

# define parameter grid for grid search
param_grid = {
    'hidden_layer_sizes': [(50,50), (100,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}

# perform a grid search with 5 fold cross-validation to find the best hyperparameters
grid_search = GridSearchCV(ann, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# print the best hyperparameters
print(grid_search.best_params_)

# initialize ANN classifier with the best hyperparameters
ann = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', solver='adam', alpha=0.0001, learning_rate='adaptive')

# fit model to training data
ann.fit(X_train, y_train)

# make predictions on testing data
y_pred = ann.predict(X_test)

# calculate evaluation metrics
ANN_acc = accuracy_score(y_test, y_pred)
ANN_acc = round(ANN_acc*100 ,2)


# print evaluation metrics
print("Accuracy Score :", ANN_acc, "%")


```
